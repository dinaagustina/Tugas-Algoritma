**2. Grafik sederhana pada Matplotlib** 

```python
#membuat grafik sederhana pada matplotlib
import matplotlib.pyplot as plt

#deklarasikan nilai x dan y
x = [1,2,3,4,5]
y = [1,8,27,64,125]

plt.plot(x,y)
#untuk meletakkan posisi titik x dan y
plt.title("Grafik y=x**3")
#ini merupakan nama dari grafik yang dihasilkan
plt.xlabel('X')
#nama untuk sumbu X
plt.ylabel('Y')
#nama untuk sumbu y
plt.show()
#menampilkan output
```

```python
#contoh grafik sederhana yaitu berupa grafik lingkaran
import matplotlib.pyplot as plt

labels = ['Dokter', 'Guru', 'Pegawai Kantor', 'Pedagang']
#merupakan label untuk setiap data
quantity = [3, 10, 12, 11]
#jumlah masing-masing data
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
#warna yang akan digunakan pada grafik

plt.title('Daftar Pekerjaan Siswa Kelas XII-A')
#nama dari grafik yang dihasilkan
plt.pie(quantity, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=90)
#plt pie digunakan untuk menyatukan labels, quantity dan colors, shadow digunakan
#untuk efek bayangan, startangle=90 yaitu nilai awal untuk tabel nilai pertama
#adalah 90 derajat

plt.axis('equal')
#memberikan aspek rasio pada grafik
plt.show()
#untuk menampilkan output grafik
```

**3. Properti pembuatan grafik pada Matplotlib**

  Ada banyak jenis properti yang dapat digunakan untuk membantu membuat grafik pada matplotlib. Beberapa sudah disebutkan dalam contoh sebelumnya. 

  Berikut adalah contoh penggunaan properti pembuatan grafik pada Matplotlib :

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(0.0, 5.0, 0.01)
#np arrange 0.0 merupakan awal dari grafik dan 3.0 merupakan akhir grafik
#0.01 merupakan perulangan untuk gelombang
s = np.sin(2.5 * np.pi * t)
#ini merupakan rumus untuk menghasilkan grafik sinus
plt.plot(t, s)
#untuk meletakkan titik-titik t dan s pada grafik

#pemberian nama pada sumbu x dan y
plt.xlabel('time (s)')
plt.ylabel('voltage (V)')

#pemberian nama grafik
plt.title('Gelombang Sinus')
#untuk menghasilkan grid jika nilainya bernilai true
plt.grid(True)

#untuk menghasilkan output berupa grafik
plt.show()
```

5. **Properti pembuatan histogram sederhana dengan Matplotlib**

```python
#input data yang diinginkan
chef = ["Arnold", "Renata", "Juna", "Jerry"]
penggemar = [99,96,98,95]

plt.figure(figsize = (10,4))
#figsize merupakan ukuran dari figura yang akan digunakan
hist = plt.bar(chef,penggemar,color = "green")
#untuk menambahkan list atau penamaan pada sumbu x dan y juga mendeklarasikan
#tambahan data seperti warna grafik
pola = ["x","o","/","-"]
#menambahkan pola pada diagram yang dihasilkan

for bar in hist :
  bar.set_hatch(pola.pop(0))
  #perulangan untuk menambahkan pola pada masing-masing diagram
plt.show()
#menunjukkan output atau grafik
```
